# Переменные
<p>В программировании слово переменная обозначает именованное место
для хранения данных, например чисел, текста, списков с числами или символами и так далее. Проще
говоря это это коробочка, на которой написано ее имя и в которую вы можете сохранить значение .
Также переменную можно рассматривать как ярлык( ссылка), которым помечены некие данные.
Значение переменной может изменяться во время выполнения программы.</p>
<p>Таким образом, в переменных можно хранить данные во время работы программы и использовать их
при вычислениях, когда они понадобятся.</p>
<p>Имя переменной называют идентификатором (от слова идентифицировать – отличать один объект от
другого). Идентификатор — это имя переменной.</p>
<p>Имена переменных в Python:</p>
<ul>
    <li>Первым символом переменной должна быть буква или знак подчеркивания(_).</li>
<li>Все символы, кроме первого, могут быть: →латинской буквой из нижнего регистра, верхнего
регистра, знаком подчеркивания или цифры(0-9).</li>
<li>Имя идентификатора не должно содержать пробелов или специальных символов(!, @, #,%, ^, &amp;, *).</li>
<li>Имя идентификатора не должно быть похоже ни на одно ключевое слово, определенное в языке.</li>
</ul>
<p>Имена идентификаторов чувствительны к регистру; например, my name и MyName не совпадают.
Желательно давать переменным «говорящие» имена, чтобы можно было сразу понять, зачем нужна
та или иная переменная. Например, переменная с именем name, скорее всего, служит для хранения
какого-то имени, а о назначении переменной abc догадаться очень сложно.</p>
<b><p>Определите, какие из следующих идентификаторов допустимы, а какие – нет.</p></b>

<p><code>123 | Vasya | @mail_ru | m11 | Петя | 1m | Митин брат | _27 | &quot;Pes barbos&quot;</code></p>

<p>В отличие от многих языков программирования (Паскаль, C, Java) переменные в языке Python не
нужно объявлять. Память для переменной выделяется автоматически тогда, когда ей присваивается
новое значение.</p>
<p>Присвоим переменной значение 5:</p>

```python
a = 5
```

<p>Знак &lt;=&gt; обозначает специальную команду – оператор присваивания, с его помощью присваивают
новое значение переменной. Он выполняется так: вычисляется выражение справа от символа «=», а
затем результат записывается в переменную, имя которой указано слева.
Оператор присваивания также позволяет изменить значение переменой:</p>

```python
name = 'Платон'
name = 'Сократ'
```

<p>Переменная может хранить только одно значение. При записи в неё нового значения «старое»
стирается, и его уже никак не восстановить. В языке Python при изменении значения переменной
выделяется новая область памяти и связывается с тем же именем
Теперь та область памяти, в которой было записано старое значение
(«Платон»), уже недоступна, потому что с ней не связано ни одно имя.</p>

# Арифметические выражения

<p>Арифметические выражения обычно записываются в одну строчку. Они могут содержать константы
(постоянные значения), имена переменных, знаки арифметических операций,
круглые скобки (для изменения порядка действий). Например, присваивание
в программе запишется как</p>

```python
b = 5
c = 2
d = 2
a = (c + b - 1) / 2 * d
```

<p>Операция умножения обозначается знаком «*», а операция деления – знаком «<code>/</code>».
Какое же действие будет выполняться первым, какое –вторым и т.д.? Это определяется приоритетом
(старшинством) операций. Они выполняются в следующем порядке:</p>
<ul>
    <li>сначала – действия в скобках;</li>
    <li>умножение и деление, слева направо;</li>
    <li>сложение и вычитание, слева направо.</li>
</ul>
<p>Таким образом, умножение и деление имеют одинаковый
приоритет, более высокий, чем сложение и вычитание.</p>
<p>Определите порядок действий компьютера при вычислении
выражения:</p>

```python
b = 5
c = 2
d = 2
a = c + b - 1 / 2 * 5
```

<p>Символы, с помощью которых выполняются математические
действия в языке Python, называются операторами. Основные математические операторы (см.
таблицу).</p>
<p>Скобки могут быть вложенными, то есть внутри скобок могут стоять еще одни скобки:</p>


```python
print(((5 + 30) * 20) / 10)
```

<p>Результат деления (операции «<code>/</code>») может быть нецелым числом, такие числа называются
вещественными. Если в переменную записать вещественное число, она будет относиться к
    типу <code>float</code>. При записи вещественных чисел в программе целую и дробную часть разделяют не запятой,
    а точкой. Например <code>x = 2.5</code></p>
    
# Операции с целыми числами

<p>Часто нужно получить целый результат деления целых чисел и остаток от деления. Например,
известен промежуток времени в секундах (скажем, 175 секунд) и нужно определить, сколько в нём
целых минут и оставшихся секунд (175 с = 2 мин 55 с). Здесь число минут – это целая часть от деления
175 на 60, а 55 секунд – это остаток от этого деления. В таких случаях в языке Python используют
    специальные операции <code>//</code> и <code>%</code> (они имеют такой же приоритет, как умножение и деление):</p>

```python
t = 175
mi = t // 60 # 2
sek= t % 60 # 55

print(mi)
print(sek)
```

<p>Символ <code>#</code> обозначает начало комментария – пояснительного текста, который не обрабатывается
транслятором. Комментарии — это пояснения для человека внутри текста программы.</p>
<p>С помощью этих операций удобно работать с отдельными цифрами числа. Остаток от деления
числа на 10 – это последняя цифра его десятичной записи.</p>

```python
N = 678
d1 = N % 10 # 8
print(d1)
```

### Чему равен остаток от деления числа N на 100?

```python
d2 = N % 100
```

<p>Если разделить число на 10 и взять только целую часть, мы «отбросим» последнюю цифру числа:
    значение <code>678 // 10</code> равно 67.</p>

```python
N = 678
d2 = N // 10 # 67
print(d2)
```

<p>В языке Python каждая переменная имеет свой тип. Тип нужен для того, чтобы определить, какие
значения может принимать переменная; какие операции можно выполнять с этой переменной; как
хранить её значения в памяти.</p>

### Определить тип переменной можно с помощью встроенной функции type:

```python
lang = "Минск"
print( type(lang) )
cost = 123
print( type(cost) )
dist = 45.678
print( type(dist) )
```

<p>Это означает, что переменные <code>lang</code>, <code>cost</code> и <code>dist</code> относятся
соответственно к типам (классам)</p>
<p><code>str</code> – символьная строка (от англ. string),</p>
<p><code>int</code> – целое число (от англ. integer),</p>
<p><code>float</code> – вещественное число (от англ. float).
<p>Транслятор Python сам определяет тип переменной по то-му значению, которое ей присваивается.</p>

<p>Для вывода значения переменной на экран используют тот же оператор <code>print</code>, который раньше

<p>Команда вывода (встроенная функция <code>print</code>) сначала автоматически преобразует числовое значение в
цепочку символов (цифр), а затем выводит эти символы так же, как обычный текст.</p>

<p>Что появится на экране после выполнения программы? Чем отличаются два оператора вывода в

```python
c = 5
print(c)
print('c')
```

<p>Что выведет на экран программма</p>

```python
a = 1
print( a )
a = 5
print( a )
```

<p>Оператор <code>i = i + 1</code> заменяет значение <code>i</code> на <code>i + 1</code>, то есть увеличивает значение переменной i на 1.
Что получится, если рассмотреть запись <code>i = i + 1</code>, как равенство – уравнение относительно переменной
i?</p>
<p>Чему будет равно значение переменной <code>i</code> после выполнения оператора <code>i = i + 1</code>, если до этого оно
было равно 17?</p>
<p>Вначале переменные имели значения <code>a = 4</code> и <code>b = 7</code>. Чему будут равны значения этих переменных
после выполнения программы?</p>

```python
a = a + 1
b = b + 1
a = a + b
b = b + a
a = a + 1
```

<p>В языке Python можно использовать сокращённую запись арифметических операций: например,
    <code>a += b</code> означает то же самое, что и <code>a = a + b</code>, а <code>a –= b</code> – то же самое, что и <code>a = a – b</code>. Особенно часто
используется увеличение переменной на единицу, которая записывается, например, так: <code>i += 1</code>.</p>
